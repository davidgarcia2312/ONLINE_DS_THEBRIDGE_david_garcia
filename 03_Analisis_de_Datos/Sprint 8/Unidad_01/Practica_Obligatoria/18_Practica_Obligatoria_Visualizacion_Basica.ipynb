```python
df_casas_cali = pd.read_csv("./data/california_cities.csv")
display(df_casas_cali)
print(df_casas_cali.info())
```

```python
df_ejer = df_casas_cali.copy()
df_ejer.dropna(inplace = True)
print(df_ejer.info())
```

### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
#1
mas_pobladas = df_ejer.sort_values(by="population_total", ascending=False).head(5)
pob_miles = mas_pobladas["population_total"] / 1000

colores = ['cornflowerblue', 'lightgrey', 'lightcoral', 'darkkhaki', 'lightpink']

plt.figure(figsize=(8, 8))
plt.bar(mas_pobladas["city"], pob_miles, color=colores)

plt.title("Ciudades ms pobladas de California", fontsize=12)
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)", fontsize=10)
plt.ylim(0, 5000)

for i, valor in enumerate(pob_miles):
    plt.text(i, valor + 100, f"{valor}", ha='center', va='bottom', fontsize=9, color='black')

plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

```python
#2
plt.figure(figsize=(8, 8))
plt.hist(df_ejer["area_total_km2"], bins=50, color="steelblue", alpha=0.6, label="Hist Area")

plt.xlabel("Superficie ($km^2$)", fontsize=20, font="Bradley Hand ITC")
plt.ylabel("Nmero de Ciudades", fontsize=20, font="Bradley Hand ITC")

plt.legend(loc='upper right', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()
```

```python
#3
df_ejer["pob_miles"] = df_ejer["population_total"] / 1000

plt.figure(figsize=(8, 8))
plt.scatter(df_ejer["area_water_km2"], df_ejer["pob_miles"], color="green", alpha=0.6, s=20)

plt.title("Poblacin por kilmetro cuadrado de agua", fontsize=12)
plt.xlabel("rea de agua en la ciudad ($km^2$)", fontsize=10)
plt.ylabel("Poblacin (miles habitantes)", fontsize=10)

outliers = df_ejer.sort_values(by="pob_miles", ascending=False).head(2)

plt.text(60, 3500, "Outliers", fontsize=10)

plt.annotate("",
             xy=(outliers.iloc[0]["area_water_km2"], outliers.iloc[0]["pob_miles"]),
             xytext=(65, 3500 + 100),
             arrowprops=dict(arrowstyle="->", color="black"))

plt.annotate("",
             xy=(outliers.iloc[1]["area_water_km2"], outliers.iloc[1]["pob_miles"]),
             xytext=(65, 3500 - 30),
             arrowprops=dict(arrowstyle="->", color="black"))

plt.tight_layout()
plt.show()
```

```python
#4
plt.figure(figsize=(8, 8))

plt.boxplot(df_ejer["elevation_m"],
            vert = False,
            patch_artist=True,
            boxprops=dict(facecolor='red'),
            medianprops=dict(color='black'),
            flierprops=dict(marker='o', alpha = 0.7))

plt.title("Diagrama de caja de la Altitud de las ciudades de California", fontsize=11)
plt.xlabel("Altitud (m)", fontsize=10)
plt.yticks([1], ["1"], fontsize=9)
plt.xticks(fontsize=9)  

plt.tight_layout()
plt.show()
```

```python
#Todo junto simplemente copiando y pegando lo de antes y posicionandolos en la matriz de graficos

fig, axs = plt.subplots(2, 2, figsize=(16, 14))

#1
plt.sca(axs[0, 0])

mas_pobladas = df_ejer.sort_values(by="population_total", ascending=False).head(5)
pob_miles = mas_pobladas["population_total"] / 1000

colores = ['cornflowerblue', 'lightgrey', 'lightcoral', 'darkkhaki', 'lightpink']

plt.bar(mas_pobladas["city"], pob_miles, color=colores)

plt.title("Ciudades ms pobladas de California", fontsize=12)
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)", fontsize=10)
plt.ylim(0, 5000)

for i, valor in enumerate(pob_miles):
    plt.text(i, valor + 100, f"{valor}", ha='center', va='bottom', fontsize=9, color='black')

plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

#2
plt.sca(axs[0,1])

plt.hist(df_ejer["area_total_km2"], bins=50, color="steelblue", alpha=0.6, label="Hist Area")

plt.xlabel("Superficie ($km^2$)", fontsize=20, font="Bradley Hand ITC")
plt.ylabel("Nmero de Ciudades", fontsize=20, font="Bradley Hand ITC")

plt.legend(loc='upper right', fontsize=10)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)

#3
plt.sca(axs[1,0])

df_ejer["pob_miles"] = df_ejer["population_total"] / 1000

plt.scatter(df_ejer["area_water_km2"], df_ejer["pob_miles"], color="green", alpha=0.6, s=20)

plt.title("Poblacin por kilmetro cuadrado de agua", fontsize=12)
plt.xlabel("rea de agua en la ciudad ($km^2$)", fontsize=10)
plt.ylabel("Poblacin (miles habitantes)", fontsize=10)

outliers = df_ejer.sort_values(by="pob_miles", ascending=False).head(2)

plt.text(60, 3500, "Outliers", fontsize=10)

plt.annotate("",
             xy=(outliers.iloc[0]["area_water_km2"], outliers.iloc[0]["pob_miles"]),
             xytext=(65, 3500 + 100),
             arrowprops=dict(arrowstyle="->", color="black"))

plt.annotate("",
             xy=(outliers.iloc[1]["area_water_km2"], outliers.iloc[1]["pob_miles"]),
             xytext=(65, 3500 - 30),
             arrowprops=dict(arrowstyle="->", color="black"))

#4
plt.sca(axs[1,1])

plt.boxplot(df_ejer["elevation_m"],
            vert = False,
            patch_artist=True,
            boxprops=dict(facecolor='red'),
            medianprops=dict(color='black'),
            flierprops=dict(marker='o', alpha = 0.7))

plt.title("Diagrama de caja de la Altitud de las ciudades de California", fontsize=11)
plt.xlabel("Altitud (m)", fontsize=10)
plt.yticks([1], ["1"], fontsize=9)
plt.xticks(fontsize=9)  

plt.tight_layout(pad = 3.0)
plt.savefig("matriz_graficos.png")
plt.show()
```
