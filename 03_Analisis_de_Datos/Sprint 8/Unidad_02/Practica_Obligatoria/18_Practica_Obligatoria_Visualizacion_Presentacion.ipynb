### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def pinta_scatter(df, col_x, col_y, col_cat=None, size=None, escala=1, mostrar_leyenda=True):

    """
    Pinta un diagrama de dispersin.

    Parmetros:
    - df: DataFrame
    - col_x: nombre de la columna numrica para el eje X
    - col_y: nombre de la columna numrica para el eje Y
    - col_cat: nombre de la columna categrica para colorear (opcional)
    - size: nombre de la columna para tamao o valor numrico fijo
    - escala: escala de los tamaos si size es una columna (por ejemplo 1/10000)
    - mostrar_leyenda: bool que controla si se muestra o no la leyenda
    """

    plt.figure(figsize=(16, 8))

    scatter_kwargs = {
        "data": df,
        "x": col_x,
        "y": col_y,
        "hue": col_cat if col_cat else None,
        "palette": "viridis" if col_cat else None,
        "alpha": 0.6,
        "edgecolor": None,
    }

    if isinstance(size, str):      #Si size es string (nombre de columna)
        scatter_kwargs["size"] = size
        scatter_kwargs["sizes"] = (10, 300)
    elif isinstance(size, (int, float)):   #Si size es numrico fijo
        scatter_kwargs["s"] = size
    elif isinstance(size, pd.Series):  # Si size es columna ya escalada o una serie numrica
        scatter_kwargs["s"] = size * escala

    sns.scatterplot(**scatter_kwargs)

    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title("Diagrama de dispersin")
    plt.tight_layout()
    plt.show()
```

En esta parte tendrs que hacer dos cosas, crear una presentacin sobre el titanic de no ms de 6 slides, sin incluir portada, con los siguientes mensajes y su apoyo en datos:  
1- El titanic fue un accidente con una tasa de mortalidad alta.  
2- En esa ocasin los datos apoyan que se dio el "Las mujeres y los nios primero", pero...  
3- ... que la clase en la que se viajes tuvo una influencia significativa y podra explicar...  
4- ... las diferencias de tasas de supervivencia entre mujeres y nios.    
5- Adems aunque el puerto de embarque parece influir, se podra asumir que lo fue el tipo de pasaje que embarc en cada uno.  

Y generar el notebook con las grficas que vas a emplear en la presentacin. Tienes que entregar ambas cosas.

```python
df_titanic = pd.read_csv("./data/titanic.csv")
```

```python
#Slide 1: El titanic fue un accidente con una tasa de mortalidad alta

supervivencia = df_titanic["survived"].value_counts()
labels = ["No sobrevivi", "Sobrevivi"]
colors = ["red", "green"]

plt.figure()
plt.pie(
    supervivencia,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 12}
)
plt.title("Supervivencia en el Titanic", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#Slide 2: En esta ocasin los datos apoyan que se dio el "Las mujeres y los nios primero", pero... 

tasa_supervivencia_quien = df_titanic.groupby("who")["survived"].mean().sort_values(ascending=False)
tasa_supervivencia_quien.plot(kind="bar", color=["green", "blue", "red"])

plt.title("Tasa de supervivencia por grupo")
plt.xlabel("Grupo de poblacin")
plt.ylabel("Tasa de supervivencia")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python

#Slide 3: ... que la clase en la que se viajes tuvo una influencia significativa y podra explicar...

# Calculamos la tasa de supervivencia por clase
tasa_supervivencia_clase = df_titanic.groupby("pclass")["survived"].mean().sort_index()

# Grfico de barras
tasa_supervivencia_clase.plot(kind="bar", color=["red", "green", "blue"])

plt.title("Tasa de supervivencia por clase")
plt.xlabel("Clase (1 = alta | 2 = media | 3 = baja)")
plt.ylabel("Tasa de supervivencia")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
# Slide 4: ... las diferencias de tasas de supervivencia entre mujeres y nios. 

filtro = df_titanic["who"].isin(["woman", "child"])
tasa_female_child = df_titanic[filtro].groupby("who")["survived"].mean()

tasa_female_child.plot(kind="bar", color=["red", "green"])

plt.title("Tasa de supervivencia: Mujeres vs Nios")
plt.ylabel("Tasa de supervivencia")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
#Slide 5: Adems aunque el puerto de embarque parece influir, se podra asumir que lo fue el tipo de pasaje que embarc en cada uno.

# Calcular tasa de supervivencia media agrupando por puerto y clase
tasa_por_puerto_y_clase = df_titanic.groupby(["embark_town", "class"])["survived"].mean().reset_index()
tasa_por_puerto_y_clase.columns = ["Puerto", "Clase", "Tasa de Supervivencia"]

sns.barplot(data=tasa_por_puerto_y_clase, x="Puerto", y="Tasa de Supervivencia", hue="Clase")

plt.title("Tasa de supervivencia por puerto de embarque y clase")
plt.ylabel("Tasa de supervivencia")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
