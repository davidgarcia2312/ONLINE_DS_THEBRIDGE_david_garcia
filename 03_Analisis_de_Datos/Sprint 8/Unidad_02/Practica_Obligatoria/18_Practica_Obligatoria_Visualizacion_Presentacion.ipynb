### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesión en vivo)

```python
def pinta_scatter(df, col_x, col_y, col_cat=None, size=None, escala=1, mostrar_leyenda=True):

    """
    Pinta un diagrama de dispersión.

    Parámetros:
    - df: DataFrame
    - col_x: nombre de la columna numérica para el eje X
    - col_y: nombre de la columna numérica para el eje Y
    - col_cat: nombre de la columna categórica para colorear (opcional)
    - size: nombre de la columna para tamaño o valor numérico fijo
    - escala: escala de los tamaños si size es una columna (por ejemplo 1/10000)
    - mostrar_leyenda: bool que controla si se muestra o no la leyenda
    """

    plt.figure(figsize=(16, 8))

    scatter_kwargs = {
        "data": df,
        "x": col_x,
        "y": col_y,
        "hue": col_cat if col_cat else None,
        "palette": "viridis" if col_cat else None,
        "alpha": 0.6,
        "edgecolor": None,
    }

    if isinstance(size, str):      #Si size es string (nombre de columna)
        scatter_kwargs["size"] = size
        scatter_kwargs["sizes"] = (10, 300)
    elif isinstance(size, (int, float)):   #Si size es numérico fijo
        scatter_kwargs["s"] = size
    elif isinstance(size, pd.Series):  # Si size es columna ya escalada o una serie numérica
        scatter_kwargs["s"] = size * escala

    sns.scatterplot(**scatter_kwargs)

    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title("Diagrama de dispersión")
    plt.tight_layout()
    plt.show()
```

## #2: Presentación

En esta parte tendrás que hacer dos cosas, crear una presentación sobre el titanic de no más de 6 slides, sin incluir portada, con los siguientes mensajes y su apoyo en datos:  
1- El titanic fue un accidente con una tasa de mortalidad alta.  
2- En esa ocasión los datos apoyan que se dio el "Las mujeres y los niños primero", pero...  
3- ... que la clase en la que se viajes tuvo una influencia significativa y podría explicar...  
4- ... las diferencias de tasas de supervivencia entre mujeres y niños.    
5- Además aunque el puerto de embarque parece influir, se podría asumir que lo fue el tipo de pasaje que embarcó en cada uno.  

Y generar el notebook con las gráficas que vas a emplear en la presentación. Tienes que entregar ambas cosas.

```python
df_titanic = pd.read_csv("./data/titanic.csv")
```

```python
#Slide 1: El titanic fue un accidente con una tasa de mortalidad alta

supervivencia = df_titanic["survived"].value_counts()
labels = ["No sobrevivió", "Sobrevivió"]
colors = ["red", "green"]

plt.figure()
plt.pie(
    supervivencia,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 12}
)
plt.title("Supervivencia en el Titanic", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#Slide 2: En esta ocasión los datos apoyan que se dio el "Las mujeres y los niños primero", pero... 

tasa_supervivencia_quien = df_titanic.groupby("who")["survived"].mean().sort_values(ascending=False)
tasa_supervivencia_quien.plot(kind="bar", color=["green", "blue", "red"])

plt.title("Tasa de supervivencia por grupo")
plt.xlabel("Grupo de población")
plt.ylabel("Tasa de supervivencia")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python

#Slide 3: ... que la clase en la que se viajes tuvo una influencia significativa y podría explicar...

# Calculamos la tasa de supervivencia por clase
tasa_supervivencia_clase = df_titanic.groupby("pclass")["survived"].mean().sort_index()

# Gráfico de barras
tasa_supervivencia_clase.plot(kind="bar", color=["red", "green", "blue"])

plt.title("Tasa de supervivencia por clase")
plt.xlabel("Clase (1 = alta | 2 = media | 3 = baja)")
plt.ylabel("Tasa de supervivencia")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
# Slide 4: ... las diferencias de tasas de supervivencia entre mujeres y niños. 

filtro = df_titanic["who"].isin(["woman", "child"])
tasa_female_child = df_titanic[filtro].groupby("who")["survived"].mean()

tasa_female_child.plot(kind="bar", color=["red", "green"])

plt.title("Tasa de supervivencia: Mujeres vs Niños")
plt.ylabel("Tasa de supervivencia")
plt.xticks(rotation=0)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
#Slide 5: Además aunque el puerto de embarque parece influir, se podría asumir que lo fue el tipo de pasaje que embarcó en cada uno.

# Calcular tasa de supervivencia media agrupando por puerto y clase
tasa_por_puerto_y_clase = df_titanic.groupby(["embark_town", "class"])["survived"].mean().reset_index()
tasa_por_puerto_y_clase.columns = ["Puerto", "Clase", "Tasa de Supervivencia"]

sns.barplot(data=tasa_por_puerto_y_clase, x="Puerto", y="Tasa de Supervivencia", hue="Clase")

plt.title("Tasa de supervivencia por puerto de embarque y clase")
plt.ylabel("Tasa de supervivencia")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
